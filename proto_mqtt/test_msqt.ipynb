```python
import paho.mqtt.client as mqtt
import time
import json
```

```python
# Callback functions 
# Callback when connecting to the broker
def on_connect(client, userdata, flags, reason_code, properties):
    if reason_code == 0:
        print("Connected successfully")
        # Subscribe to topics
        client.subscribe("test/topic2")
    else:
        print(f"Failed to connect, reason code: {reason_code}")

# Callback when receiving a message
def on_message(client, userdata, msg):
    print(f"Message received on topic {msg.topic}: {msg.payload.decode()}")
```

```python
# Create client instance
client = mqtt.Client(mqtt.CallbackAPIVersion.VERSION2)

# Set callbacks
client.on_connect = on_connect
client.on_message = on_message
```

```python
broker_address = "test.mosquitto.org"
#TODO add local broker address
port = 1883
```

```python
try:
    client.connect(broker_address, port, keepalive=60)
    
    # Start the network loop in a separate thread
    client.loop_start()
    
    # Publish some messages
    for i in range(40):
        message = f"Hello MQTT! Message #{i+1}"
        client.publish("test/topic2", message)
        print(f"Published: {message}")
        time.sleep(1)
    
    # Keep running for a while to receive messages
    time.sleep(1)
    
    # Clean disconnect
    client.loop_stop()
    client.disconnect()
    
except Exception as e:
    print(f"Error: {e}")
```
